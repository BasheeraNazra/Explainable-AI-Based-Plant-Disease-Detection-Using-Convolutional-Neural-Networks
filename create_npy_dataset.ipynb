```python
import numpy as np
import os
import cv2
```

```python
#Dataset folder path
```

```python
dataset_path = "PlantVillage"
```

```python
#Class names (15 classes)
```

```python
plants = [
'Pepper__bell___Bacterial_spot',
'Pepper__bell___healthy',
'Potato___Early_blight',
'Potato___healthy',
'Potato___Late_blight',a
'Tomato_Bacterial_spot',
'Tomato_Early_blight',
'Tomato_healthy',
'Tomato_Late_blight',
'Tomato_Leaf_Mold',
'Tomato_Septoria_leaf_spot',
'Tomato_Spider_mites_Two_spotted_spider_mite',
'Tomato__Target_Spot',
'Tomato__Tomato_mosaic_virus',
'Tomato__Tomato_YellowLeaf__Curl_Virus'
]
```

```python
# Lists to store images and labels
```

```python
X = []
Y = []
```

```python
print("Starting dataset conversion...\n")

# ================================
# Loop through each class folder
# ================================
for i in range(len(plants)):

    class_folder = os.path.join(dataset_path, plants[i])

    print("Processing class:", plants[i])

    for image_name in os.listdir(class_folder):

        image_path = os.path.join(class_folder, image_name)

        try:
            # Read image
            img = cv2.imread(image_path)

            # Resize image
            img = cv2.resize(img, (64,64))

            # Convert to numpy array
            img_array = np.array(img)

            # Append to dataset
            X.append(img_array)
            Y.append(i)

        except:
            print("Skipped image:", image_path)

print("\nTotal images processed:", len(X))
```

```python
# Convert lists to numpy arrays
```

```python
X = np.array(X)
Y = np.array(Y)
```

```python
print("Image dataset shape:", X.shape)
print("Label dataset shape:", Y.shape)
```

```python
#Save as .npy files
```

```python
np.save("Model/myimg_data.txt", X)
np.save("Model/myimg_label.txt", Y)

print("\nDataset successfully saved as:")
print("myimg_data.txt.npy")
print("myimg_label.txt.npy")
```
